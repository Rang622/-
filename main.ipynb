```python
# Management Program

import pickle, os

# 공용 변수
choice = option = search_option = index = arr_index = 0
what_office = what_area = 0

# 개인별 변수 및 리스트
office_key = area_key =  supply_key = 0

office = []
office_code = build_name = office_num = mgr = None

area = []
area_code=area_name=area_human = None

supply = []
supply_key = office_code = area_code = supply_type = supply_name = supply_price = rent_name = rent_date = return_date = None

# List for Pickling
total =[office, area, supply]

# Edit Options: Search, Input, Update, Delete
def edit_option():
    print('_________________________________________________________________________________') 
    print('1.조회')
    print('2.입력') 
    print('3.수정') 
    print('4.삭제')
    if choice == 3:
        print('5.대여')
        print('6.반납')
        print('7.이관')
    print('_________________________________________________________________________________')

    # choose an edit option
    option = int(input('> '))
    return option
    
# Range Warning
def range_warning(low, high):
    print('\nPlease enter a number between ', low, 'to ', high, '!\n')
    
# Print All: Office, Area, Supply
def print_all(list_name):
    # 사무실 관리
    if list_name == office:
        print('사무실코드 건물명칭\t 사무실호실 사무실담당자')
        print('--------------------------------------------------------------------------------')
        for i in range(len(office)):
            print(office[i][0], '\t', office[i][1], '\t', office[i][2], '\t', office[i][3], '\t') 
        print('')
        
    # 영역 관리
    elif list_name == area:
        print('사무실코드 영역코드\t 영역명칭\t 영역담당자')
        print('--------------------------------------------------------------------------------')
        for i in range(len(area)):
            print(area[i][0], '\t', area[i][1], area[i][2], '\t', area[i][3], '\t')
        print('')
        
    # 물품 관리
    elif list_name == supply:
        print('물품코드\t 사무실코드 영역코드\t 물품종류\t 물품명칭\t 물품가격\t 대여자\t 대여일자\t 반납일자')
        print('--------------------------------------------------------------------------------')
        for i in range(len(supply)): 
            print(supply[i][0], '\t', supply[i][1], '\t', supply[i][2], '\t',
                  supply[i][3], '\t', supply[i][4], '\t', supply[i][5], '\t',
                  supply[i][6], supply[i][7], supply[i][8])
        print('')
        
    else: print('Not a valid list')
    
# Print Selected: Office, Area, Supply
def print_selected(list_name,index):
    # 사무실 관리
    if list_name == office:
        print('사무실코드 >', office[index][0])
        print('건물명칭 \t >', office[index][1])
        print('사무실호실  >', office[index][2])
        print('사무실담당자 >', office[index][3])
        print('')
        
    # 영역 관리
    if list_name == area:
        print('사무실코드 \t >', area[index][0])
        print('영역코드 \t >', area[index][1])
        print('영역명칭 \t >', area[index][2])
        print('영역담당자 \t >', area[index][3])
        print('')
        
    # 물품 관리
    if list_name == supply:
        print('물품코드 \t >', supply[index][0])
        print('사무실코드 \t >', supply[index][1])
        print('영역코드 \t >', supply[index][2])
        print('물품종류 \t >', supply[index][3])
        print('물품명칭 \t >', supply[index][4])
        print('물품가격 \t >', supply[index][5])
        print('대여자 \t >', supply[index][6])
        print('대여일자 \t >', supply[index][7])
        print('반납일자 \t >', supply[index][8])
        print('')

# Get Index
def get_index(arr):
    while True:
        if arr == office:
            arr_index = int(input('사무실코드 > '))
        elif arr == area:
            arr_index = int(input('영역코드 \t >'))
        elif arr == supply:
            arr_index = int(input('Index \t > ')) - 1
        
        if arr_index in range(len(arr)): 
            return arr_index
            break
        else: range_warning(0,len(arr)-1)
            
# Get Office Code
def get_office_code():
    while True:
        # print all office
        print_all(office)
        #get office code
        what_office = int(input('사무실코드 \t > '))
        if what_office in range(len(office)): break
        else: range_warning(len(office))
    return what_office

# Get Office Code & Area Code
def get_office_area_code():
    while True:
        # print all office
        print_all(office)
        # get office code
        what_office = int(input('사무실코드 \t > '))
        if int(what_office) in range(len(office)):
            while True:
                # print all area of the office
                print('사무실코드\t 영역코드\t 영역명칭\t 영역담당자\t')
                print('--------------------------------------------------------------------------------')
                for i in range(len(area)):
                    if int(area[i][0]) == what_office:
                        print(area[i][0], '\t', area[i][1], '\t', area[i][2], '\t', area[i][3], '\t')
                    i += 1
                print('')
                # get area code
                what_area = int(input('영역코드 \t > '))
                if what_area in range(len(area)): break
                else: range_warning(0, len(area)-1)
            break
        else: range_warning(0, len(office)-1)
    return what_office, what_area

# Print all area in an office
def area_in_office(what_office):
    if len(area) == 0: print('No area exist.\n')
    else:
        print('사무실코드 영역코드 영역명칭\t 영역담당자\t')
        print('--------------------------------------------------------------------------------')
        for i in range(len(area)):
            if area[i][0] == what_office:
                print(area[i][0], '\t', area[i][1], '\t', area[i][2], '\t', area[i][3])
        print('')
        
# Print all supplies in an area
def supply_in_area(what_office, what_area):
    if len(supply) == 0: print('No supply exist.\n')
    else:
        print('Index 물품코드 사무실코드 영역코드 물품종류\t 물품명칭\t 물품가격\t 대여자\t 대여일자\t 반납일자\t')
        print('--------------------------------------------------------------------------------')
        for i in range(len(supply)):
            if what_office == supply[i][1] and what_area == supply[i][2]:
                print(i+1, '\t',
                      supply[i][0], '\t', supply[i][1], '\t', supply[i][2], '\t', 
                      supply[i][3], '\t', supply[i][4], '\t', supply[i][5], '\t', 
                      supply[i][6], '\t', supply[i][7], supply[i][8], '\t')
        print('')

# Main
while True: 
    # Option
    print('*********************************************************************************') 
    print('1.사무실 관리')
    print('2.영역 관리') 
    print('3.물품 관리') 
    print('4.검색') 
    print('5.저장')
    print('6.읽어오기')
    print('7.종료')
    print('*********************************************************************************')
    
    # Get Choice
    choice = int(input('> '))
    
    if choice == 1:
        print('\n사무실 관리')
        option = edit_option()
        
        # 1.1.Search
        if option == 1: 
            if len(office) == 0: 
                print('There is no office.')
                print('Add an office.\n')
            else: print_all(office)
            
        # 1.2.Input
        elif option == 2:
            # get all values of office
            office_code = office_key
            office_key += 1
            build_name = input('건물명칭 \t >')
            office_num = input('사무실호실 \t >')
            mgr = input('사무실담당자 \t>')
            print('')
            
            # append on office
            office.append([office_code, build_name, office_num, mgr])

        # 1.3.Update
        elif option == 3: 
            print_all(office)
            arr_index = get_index(office)
            
            # update info
            build_name = input('건물명칭 \t >')
            office_num = input('사무실호실 \t >')
            mgr = input('사무실담당자 \t> ')
            
            # append on office
            office[arr_index] = [office[arr_index][0], build_name, office_num, mgr]
            
            # check update
            print('\nSucessfully Updated.\n')
            print_all(office)
            
        # 1.4.Delete
        elif option == 4:
            print_all(office)
            arr_index = get_index(office)
            
            # remove info
            for i in range(len(area)):
                if area[i][0] == office[arr_index][0]:
                    for j in range(len(supply)):
                        if supply[j][2] == area[arr_index][1]: del supply[j]
                    del area[i]
            del office[arr_index]
            
            # check delete
            print('\nSucessfully Deleted.\n')
            if len(office) == 0: print('There is no data.')
            else: print_all(office)
                
    if choice == 2: 
        print('\n영역 관리')
        option = edit_option()
        
        # 2.1.Search
        if option == 1:
            if len(office) == 0: 
                print('There is no office.')
                print('Add an office first.\n')
            else:
                what_office=get_office_code()
                area_in_office(what_office)
            
        # 2.2.Input
        if option == 2:
            what_office=get_office_code()
            
            # get all values of area
            office_code = office[what_office][0]
            area_code = area_key
            area_key += 1
            area_name=input('영역명칭 \t>')
            area_human=input('영역담당자 \t>')
            print('')

            # append on area
            area.append([office_code, area_code, area_name, area_human])
                    
        # 2.3.Update
        if option == 3:
            what_office=get_office_code()
            area_in_office(what_office)
            
            # get area index
            arr_index = get_index(area)
            
            # update info
            office_code=what_office
            area_name = input('영역명칭 \t>')
            area_human=input('영역담당자 \t>')
        
            # append on area
            area[arr_index] = [office_code, area[arr_index][1], area_name, area_human]

            # check update
            print('\nSucessfully Updated.\n')
            area_in_office(what_office)
            
        # 2.4.Delete
        if option == 4:
            what_office=get_office_code()
            area_in_office(what_office)
            
            #get area index
            arr_index = get_index(area)
            
            
            # remove info
            for i in range(len(supply)):
                if supply[i][2] == area[arr_index][1]: del supply[i]
            del area[arr_index]
            
            # check delete
            print('\nSucessfully Deleted.\n')
            if len(area) == 0: print('There is no data.')
            else: area_in_office(what_office)

    if choice == 3:
        print('\n물품 관리')
        option = edit_option()
        
        # 3.1.Search
        if option == 1:
            if len(office) == 0: 
                print('There is no office.')
                print('Add an office first.\n')
            elif len(area) == 0:
                print('There is no area.')
                print('Add an area first.\n')
            else:
                what_office, what_area = get_office_area_code()
                supply_in_area(what_office, what_area)

        # 3.2.Input
        if option == 2:
            what_office, what_area = get_office_area_code()
            
            # get all values of supply
            supply_key = input('물품코드 \t > ')
            office_code = office[what_office][0]
            area_code = area[what_area][1]
            supply_type = input('물품종류 \t > ')
            supply_name = input('물품명칭 \t > ')
            supply_price = input('물품가격 \t > ')
            print('')
            
            # append on supply
            supply.append([supply_key, office_code, area_code, 
                           supply_type, supply_name, supply_price, 
                           None, None, None])

        # 3.3.Update
        if option == 3:
            what_office, what_area = get_office_area_code()
            supply_in_area(what_office, what_area)
            
            # get supply index
            arr_index = get_index(supply)

            # update info
            supply_key = input('물품코드 \t > ')
            office_code = what_office
            area_code = what_area
            supply_type = input('물품종류 \t > ')
            supply_name = input('물품명칭 \t > ')
            supply_price = input('물품가격 \t > ')
            
            # append on supply
            supply[arr_index] = [supply_key, office_code, area_code, 
                                 supply_type, supply_name, supply_price,
                                 supply[arr_index][6], supply[arr_index][7], supply[arr_index][8]]
            
            # check update
            print('\nSucessfully Updated.\n')
            supply_in_area(what_office, what_area)
            
        # 3.4.Delete
        if option == 4:
            what_office, what_area = get_office_area_code()
            supply_in_area(what_office, what_area)
            
            # get supply index
            arr_index = get_index(supply)
            
            # remove info
            del supply[arr_index]
            
            # check delete
            print('\nSuccessfully Deleted.\n')
            if len(supply) == 0: print('There is no data.')
            else: supply_in_area(what_office, what_area)
                
        # 3.5.Rent
        if option == 5:
            what_office, what_area = get_office_area_code()
            supply_in_area(what_office, what_area)
            
            # get supply index
            arr_index = get_index(supply)
            
            # rent a supply
            rent_name = input('대여자 \t > ')
            rent_date = input('대여일자 \t > ')

            
            # append on supply
            supply[arr_index] = [supply[arr_index][0], supply[arr_index][1], supply[arr_index][2], 
                                 supply[arr_index][3], supply[arr_index][4], supply[arr_index][5],
                                 rent_name, rent_date, None]
                
            # check rent
            print('\nSuccessfully Rented.\n')
            supply_in_area(what_office, what_area)
            
        # 3.6.Return
        if option == 6:
            what_office, what_area = get_office_area_code()
            supply_in_area(what_office, what_area)

            # get supply index
            arr_index = int(input('Index \t > ')) - 1

            # rent a supply
            return_date = input('반납일자 \t > ')

            # append on supply
            supply[arr_index] = [supply[arr_index][0], supply[arr_index][1], supply[arr_index][2], 
                                 supply[arr_index][3], supply[arr_index][4], supply[arr_index][5],
                                 supply[arr_index][6], supply[arr_index][7], return_date]

            # check rent
            print('\nSuccessfully Returned.\n')
            supply_in_area(what_office, what_area)
            
        # 3.7.Transfer
        if option == 7:
            while True:
                what_office, what_area = get_office_area_code()
                supply_in_area(what_office, what_area)

                # get supply index
                arr_index = get_index(supply)

                # transfer a supply
                print_all(office)
                office_code = int(input('옮길 사무실코드 > '))
                if office_code in range(len(office)):
                    print('')
                    area_in_office(office_code)
                    area_code = int(input('옮길 영역코드 > '))
                    if area_code in range(len(area)):
                        break
                    else: print('Not available area code.')
                else: print('Not available office code.')

            # append on supply
            supply[arr_index] = [supply[arr_index][0], office_code, area_code, 
                                 supply[arr_index][3], supply[arr_index][4], supply[arr_index][5],
                                 supply[arr_index][6], supply[arr_index][7], supply[arr_index][8]]

            # check transfer
            print('\nSuccessfully Transferred.\n')
            supply_in_area(office_code, area_code)
            
    # 4.검색
    if choice == 4: 
        print('\n검색')

        # print search options
        print('_________________________________________________________________________________')
        print('1.물품 전체 조회')
        print('2.물품 코드 조회')
        print('_________________________________________________________________________________')

        # get search index
        while True:
            search_option = int(input('> '))
            if search_option == 1 or search_option == 2: break
            else: 
                print(search_option, ' is not available.')
                print('Choose between 1 and 2.\n')
        
        # 4.1.inquire all supplies
        if search_option == 1 : 
            for i in range(len(office)):
                print('<#', office[i][2], ' ', office[i][1], '>')
                print('_________________________________________________________________________________')
                for j in range(len(area)):
                    if area[j][0] == office[i][0]:
                        print('*', area[j][2], ': ', end='')
                        for k in range(len(supply)):
                            if area[j][1] == supply[k][2]:
                                print('(#', supply[k][0], ')', supply[k][4], '\t', end='')
                        print('')
                print('*********************************************************************************')

        # 4.2.inquire supply codes
        if search_option == 2 :
            print('Index \t 물품코드\t 물품명칭')
            print('--------------------------------------------------------------------------------')
            for i in range(len(supply)): 
                print(i+1, '\t', supply[i][0], '\t', supply[i][4])
            print('')

            # get supply_code index
            arr_index = int(input('Index \t > ')) - 1

            #print_selected(supply, arr_index)
            print('물품코드 사무실코드 영역코드 물품종류 물품명칭 물품가격 대여자 대여일자 반납일자') 
            print('--------------------------------------------------------------------------------')
            print(supply[arr_index][0], '\t', supply[arr_index][1], '\t', supply[arr_index][2], '\t',
                  supply[arr_index][3], '\t', supply[arr_index][4], '\t', supply[arr_index][5], '\t',
                  supply[arr_index][6], '\t', supply[arr_index][7], '\t', supply[arr_index][8])

    # 5.저장 
    if choice == 5:
        print('\n저장')
        
        # save data: output.txt
        os.path.isfile('list.dat')
        #save
        file = open('list.dat','wb')
        pickle.dump(total, file)
        file.close()
        print('Sucessfully Saved.')

    # 6.읽어오기
    if choice == 6:
        print('\n읽어오기')
        
        # open data: output.txt
        with open('list.dat', 'rb') as file: load_file = pickle.load(file)
        
        # check read
        print('\nSucessfully Read.')
        print('Stored data :')
        print('--------------------------------------------------------------------------------------------------')
        print(load_file)
        #load_file.close()
    
    # 7.종료
    if choice == 7:
        print('\n프로그램을 종료합니다.')
        break
    
    # 예외 처리
    if choice not in range(1,8): range_warning(1,7)
        
    # make chart neat
    print('')
```
